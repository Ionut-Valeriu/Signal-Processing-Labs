## Exercițiul 1

Fie semnalele continue $x(t) = \cos (520 \pi t + \pi/3)$, $y(t) = \cos (280 \pi t - \pi/3)$ și $z(t) = \cos (120 \pi t + \pi/3)$.

- În Python, simulați axa reală de timp printr-un șir de numere suficient de apropiate, spre exemplu $[0:0.0005:0.03]$ (adică intervalul $[0, 0.03]$, discretizat cu un pas de $0.0005$).

```python
min = 0
max = 0.03
pas = 0.0005

t = np.arange(min, max, pas)

x = np.cos(520*PI*t + PI/3)
y = np.cos(280*PI*t - PI/3)
z = np.cos(120*PI*t + PI/3)
```

- Construiți semnalele $x(t)$, $y(t)$ și $z(t)$ și afișați-le grafic, în câte un subplot.

```python
fig, axs = plt.subplots(3)
axs[0].plot(t, x)
axs[1].plot(t, y)
axs[2].plot(t, z)
```

- Eșantionați semnalele cu o frecvență de 200 Hz pentru a obține $x[n]$, $y[n]$ și $z[n]$ și afișați-le grafic, în câte un subplot.

```python
min = 0
max = 0.03
fr = 200

t = np.arange(min, max, 1/fr)

x = np.cos(520*PI*t + PI/3)
y = np.cos(280*PI*t - PI/3)
z = np.cos(120*PI*t + PI/3)

fig, axs = plt.subplots(3)
axs[0].plot(t, x)
axs[1].plot(t, y)
axs[2].plot(t, z)
```

- Un semnal **sinusoidal** de frecvență 400 Hz, care să conțină 1600 de eșantioane.

```python
samples = 1600
start = 0
stop = 0.03

fr = 400

# t = np.arange(start, stop, 1/samples)
t = np.linspace(start, stop, samples)

x1 = np.sin(2*PI*t*fr)

fig = plt.plot(t, x1)
```

- Un semnal **sinusoidal** de frecvență 800 Hz, care să dureze 3 secunde.

```python
start = 0
stop = 3

fr = 800

# t = np.arange(start, stop, 1/fr)
t = np.linspace(start, stop, fr*2)

x2 = np.sin(2*PI*t*fr)

fig = plt.plot(t, x2)
```

- Un semnal de tip **sawtooth** de frecvență 240 Hz (puteți folosi funcțiile `numpy.floor` sau `numpy.mod`).

```python
start = -1
stop = 1

fr = 240

# t = np.arange(start, stop, 1/fr)
t = np.linspace(start, stop, fr)

x3 = 2*(t - np.floor(t+1/2))
x4 = np.mod(t, 1)


fig, axs = plt.subplots(2)
axs[0].plot(t, x3)
axs[1].plot(t, x4)
```

- Un semnal de tip **square** de frecvență 300 Hz (puteți folosi funcția `numpy.sign`).

```python
start = -1
stop = 1

fr = 300

# t = np.arange(start, stop, 1/fr)
t = np.linspace(start, stop, fr)

x5 = 4*np.floor(t)-2*np.floor(2*t) + 1
x6 = np.sign(np.sin(2*PI*fr*t))
x7 = np.sign(np.cos(2*PI*fr*t))

fig, axs = plt.subplots(3)
axs[0].plot(t, x5)
axs[1].plot(t, x6)
axs[2].plot(t, x7)
```
